```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('spambase-train.csv')
testdata = pd.read_csv('spambase-test.csv')
```

```python
Y = data[data.columns[-1]].as_matrix()
```

```python
testY = testdata[testdata.columns[-1]].as_matrix()
```

```python
Xframe = data.drop(data.columns[-1], axis=1)
X = Xframe.as_matrix()
X = np.append(np.ones((X.shape[0],1)),X,axis=1)
```

```python
testXframe = testdata.drop(data.columns[-1],axis=1)
testX = testXframe.as_matrix()
testX = np.append(np.ones((testX.shape[0],1)),testX,axis=1)
```

```python
def sigmoid(x):
    g = 1 / (1+np.exp(-x))
    return g
```

```python
w = np.zeros(X.shape[1])
lr = 1;
reg = 1;
m = len(Y)
```

```python
gradient = np.array([1,2])
while np.sqrt(gradient.dot(gradient)) > .001:
    a = sigmoid(np.dot(X,w))
    cost = (1/m) * sum((-Y * np.log(a)) - ((1-Y) * np.log(1-a))) + (reg/(2*m))*sum(np.square(w))
    #b = b - (1/m)*sum((a-Y))
    gradient = np.sum(np.multiply((a-Y),X.T),axis=1)/m + (reg/m)*w
    w = w - (lr*gradient)
```

```python
index = 0
print('bias :', w[0])
for wi in w[1:]:
    print(data.columns[index], ': ', wi)
    index+=1
```

```python
def getAccuracy():
    prediction = sigmoid(np.dot(X,w))
    prediction = prediction>=.5
    trainAccuracy = len(Y[Y==prediction])/float(len(Y))
    return trainAccuracy

def getTestAccuracy():
    testPred = sigmoid(np.dot(testX,w))
    testPred = testPred>=.5
    testAccuracy = testY[testY==testPred].shape[0]/float(testY.shape[0])
    return testAccuracy
```

```python
print('Training Accuracy: ', getAccuracy())
print('Test Accuracy: ',getTestAccuracy())
```
